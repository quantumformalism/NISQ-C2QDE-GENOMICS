```python
# Packages

import numpy as np
import matplotlib.pyplot as plt
```

#Boltzmann Distribution: $p(E_{i})\propto e^{-E_{i}/k_{B}T}$

```python
# Defining the Boltzmann distribution

def boltzmann(x, kT):
    return np.exp(-E/kT)
```

```python
# Parameter values for the various graphs

kT_values = [1, 2, 3, 4] # Lumping the product K*T together
```

```python
# Graph / subplot parameterss 

fig, axs = plt.subplots(1, len(kT_values), figsize = (12,4))

########################################################################

# Plotting the graphs

for i, kT in enumerate(kT_values):
    E = np.linspace(0, 10, 1000)
    y = boltzmann(E, kT)
    axs[i].plot(E, y)
    axs[i].set_xlabel('E')
    axs[i].set_ylabel('P(E)')
    axs[i].set_title(f'kT = {kT}')

########################################################################

# Display the graphs

plt.tight_layout()
plt.savefig('boltzmann_plots.pdf') # Saving the graphs in PDF format
plt.show()



```
