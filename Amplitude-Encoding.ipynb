# Amplitude Enoding

```python
import numpy as np

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, execute
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator
from qiskit.visualization import plot_state_city, plot_bloch_multivector
from qiskit.visualization import plot_state_paulivec, plot_state_hinton
from qiskit.visualization import plot_state_qsphere

# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

### Example 1:

```python
# function to convert a DNA sequence to a binary string
def dna_to_binary(dna_seq):
    dna_dict = {'A': '00', 'C': '01', 'G': '10', 'T': '11'}
    return ''.join([dna_dict[base] for base in dna_seq])

# input DNA sequence
dna_seq = 'ACTGACGTAGC'

# convert DNA sequence to binary string
binary_str = dna_to_binary(dna_seq)

# create quantum circuit with n qubits, where n is the length of the binary string
n = len(binary_str)
qc = QuantumCircuit(n)

# apply amplitude encoding to quantum circuit
for i, bit in enumerate(binary_str):
    if bit == '1':
        qc.x(i)
qc.h(range(n))

# apply rotation gates to the qubits
theta = 2*np.arcsin(1/np.sqrt(2**n))
for i in range(n):
    qc.rz(theta, i)

# execute the quantum circuit on a simulator
backend = Aer.get_backend('statevector_simulator')
job = execute(qc, backend)
result = job.result()

# get the statevector from the result and print it
statevector = result.get_statevector(qc)
print(statevector)
```
